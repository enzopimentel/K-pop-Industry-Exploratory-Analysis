```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://dbkpop.com/db/k-pop-girlgroups/"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table')

headers = []
for column in table.find_all('th'):
    title = column.get_text(strip=True)
    headers.append(title)

rows = []
for row in table.find_all('tr'):
    cells = []
    for cell in row.find_all('td'):
        cells.append(cell.text.strip())
    rows.append(cells)
```

```python
headers
```

```python
rows
```

```python
df = pd.DataFrame(rows, columns=headers)
```

```python
url = "https://dbkpop.com/db/k-pop-boybands/"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table')

headers = []
for column in table.find_all('th'):
    title = column.get_text(strip=True)
    headers.append(title)

b_rows = []
for b_row in table.find_all('tr'):
    cells = []
    for cell in b_row.find_all('td'):
        cells.append(cell.text.strip())
    b_rows.append(cells)
```
